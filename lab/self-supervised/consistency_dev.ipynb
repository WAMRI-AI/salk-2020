# Model

```python
class ContrastiveCallback(LearnerCallback):
    def on_batch_begin(self, last_input, last_target, **kwargs): 
        xb_pos = last_input # (bs x w x h)
        xb_tar = last_target
        xb = (torch.cat([xb_pos, xb_tar]),) # (bs*2 x w x h)
        return {'last_input': xb, 'last_target': torch.tensor([1.])}
```

```python
class CustomLoss(nn.Module):
    def __init__(self, tau=0.1):
        super(CustomLoss, self).__init__()
        # temperature parameter
        self.tau = tau
    
    def normalize(self, x):
        """Normalize tensor using L2"""
        norm = torch.norm(x, p=2, dim=1).unsqueeze(1)
        return x.div(norm)
    
    def pairwise_dists(self, bs, samples):
        return F.cosine_similarity(samples, samples.view(bs, 1, -1), dim=-1) / self.tau
    
    def sum_dists(self, S):
        """This computes the denominator of NT-Xent loss"""
        # exponentiate 
        exp_S = torch.exp(S)
        return exp_S.sum(axis=1) - (e**(1. / self.tau))  # bs x bs
    
    def forward(self, pred, target):
        bs = pred.shape[0]
        pred = self.normalize(pred)
        # pos_dists
        S = self.pairwise_dists(bs, pred)  # (bs,bs)
        # summed_dists per target
        summed_exp_dists = self.sum_dists(S)  # (bs,)
        loss = torch.log(summed_exp_dists) - S
        return loss.mean()
```

```python
class Head(nn.Module):
    def __init__(self, in_dim=512, out_dim=128):
        super(Head, self).__init__()
        self.lin1 = nn.Linear(in_dim, out_dim)
        self.act = nn.ReLU()
        self.lin2 = nn.Linear(out_dim, out_dim)
    
    def forward(self, x):
        x = self.lin1(x)
        x = self.act(x)
        return self.lin2(x)
```

```python
from fastai.vision import *
def get_patched_data(data_pth, hr_dir, bs, in_sz, out_sz,
                     tfms=None, num_workers=4, noise=None, max_zoom=1.1):
    src = get_patched_src(data_pth, hr_dir)
    
    data = (src
                .transform(tfms, size=in_sz, resize_method=ResizeMethod.CROP)
                .transform_y(None, size=out_sz, resize_method=ResizeMethod.CROP)
                .databunch(bs=bs, num_workers=num_workers)
                .normalize(imagenet_stats, do_y=True))
    data.c = 3
    data.train_ds.tfms_y = None
    return data


def get_patched_src(data_pth, hr_dir):
    hr_tifs = data_pth/f'{hr_dir}'

    def map_to_hr(x):
        return hr_tifs/x.relative_to(hr_tifs).with_suffix(".tif")

    src = (ImageImageList
            .from_folder(hr_tifs)
            .split_by_rand_pct()
            .label_from_func(map_to_hr))
    return src
```
